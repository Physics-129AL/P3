```python
# Problem 1 B

import matplotlib.pyplot as plt
import random
import time

def sort_using_list(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

def sort_using_dict(arr):
    element_counts = {}
    for element in arr:
        if element in element_counts:
            element_counts[element] += 1
        else:
            element_counts[element] = 1
    
    sorted_elements = []
    for element, count in sorted(element_counts.items()):
        sorted_elements.extend([element] * count)
    
    return sorted_elements

def generate_random_list(n):
    return [random.randint(1, 10000) for _ in range(n)]

# Test different input sizes
input_sizes = [100, 500, 1000, 5000, 10000]
execution_times_list = []
execution_times_dict = []

for size in input_sizes:
    input_list = generate_random_list(size)
    
    list_copy = input_list.copy()
    start_time = time.time()
    sort_using_list(list_copy)
    end_time = time.time()
    execution_time_list = end_time - start_time
    execution_times_list.append(execution_time_list)
    
    dict_copy = input_list.copy()
    start_time = time.time()
    sort_using_dict(dict_copy)
    end_time = time.time()
    execution_time_dict = end_time - start_time
    execution_times_dict.append(execution_time_dict)

plt.figure(figsize=(15, 5))

# Time Complexity comparison
plt.subplot(1, 2, 1)
plt.plot(input_sizes, execution_times_list, marker='o', color='blue', label='Using List')
plt.plot(input_sizes, execution_times_dict, marker='o', color='orange', label='Using Dictionary')
plt.title("Time Complexity Comparison")
plt.xlabel("Input Size")
plt.ylabel("Execution Time (seconds)")
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.legend()
plt.grid(True)

# Time Complexity difference
time_difference = [dict_time - list_time for dict_time, list_time in zip(execution_times_dict, execution_times_list)]
plt.subplot(1, 2, 2)
plt.plot(input_sizes, time_difference, marker='o', color='green')
plt.title("Time Complexity Difference")
plt.xlabel("Input Size")
plt.ylabel("Execution Time Difference (seconds)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Problem 1 C
import numpy as np
import scipy.optimize as sp

def exponential_func(n, T0, n0):
    return T0 * np.exp(-n / n0)

params_list, covariance_list = sp.curve_fit(exponential_func, input_sizes, execution_times_list)
params_dict, covariance_dict = sp.curve_fit(exponential_func, input_sizes, execution_times_dict)

T0_list, n0_list = params_list
T0_dict, n0_dict = params_dict

print(f"List Time Complexity: T(n) = {T0_list:.2f} * e^(-n/{n0_list:.2f})")
print(f"Dictionary Time Complexity: T(n) = {T0_dict:.2f} * e^(-n/{n0_dict:.2f})")

if n0_list > n0_dict:
    print("The decay constant for the list time complexity is larger.")
else:
    print("The decay constant for the dictionary time complexity is larger.")

print("This implies that the list-based sorting algorithm has a faster decay rate as the input size increases.")
```

```python
# Problem 1 D
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import random
import time

# Define the existing sorting functions
def numpy_sort(arr):
    return np.sort(arr)

def python_sorted(arr):
    return sorted(arr)

def list_sort(arr):
    arr.sort()
    return arr

def scipy_sort(arr):
    return sp.sort(arr)

# Define the sorting functions you provided
def sort_using_list(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def sort_using_dict(arr):
    element_counts = {}
    for element in arr:
        if element in element_counts:
            element_counts[element] += 1
        else:
            element_counts[element] = 1
    
    sorted_elements = []
    for element, count in sorted(element_counts.items()):
        sorted_elements.extend([element] * count)
    
    return sorted_elements

# Generate random list
def generate_random_list(n):
    return [random.randint(1, 10000) for _ in range(n)]

# Define input sizes
input_sizes = [100, 500, 1000, 5000, 10000]
```

```python
# 1 D

# Define the sorting methods and their labels
sorting_methods = [
    numpy_sort,
    python_sorted,
    list_sort,
    scipy_sort,
    sort_using_list,
    sort_using_dict
]

labels = [
    "numpy.sort()",
    "sorted()",
    ".sort()",
    "scipy.sort()",
    "Your Sort Method (List)",
    "Your Sort Method (Dict)"
]

execution_times = []

for method in sorting_methods:
    execution_times_method = []
    for size in input_sizes:
        input_list = generate_random_list(size)
        
        start_time = time.time()
        method(input_list)
        end_time = time.time()
        
        execution_time = end_time - start_time
        execution_times_method.append(execution_time)
    
    execution_times.append(execution_times_method)
```

```python
# 1 D
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
for i, times in enumerate(execution_times):
    plt.plot(input_sizes, times, marker='o', label=f"{i+1}: {labels[i]}")

plt.title("Execution Times for Different Sorting Methods")
plt.xlabel("Input Size")
plt.ylabel("Execution Time (seconds)")
plt.yscale('log') 
plt.legend()
plt.grid(True)

decay_constants = []

for times in execution_times:
    n = np.array(input_sizes)
    log_times = np.log(times)
    A = np.vstack([n, np.ones(len(n))]).T
    m, c = np.linalg.lstsq(A, log_times, rcond=None)[0]
    decay_constants.append(m)

plt.subplot(1, 2, 2)
plt.barh(labels, decay_constants, color='skyblue')
plt.title("Decay Constants for Different Sorting Methods")
plt.xlabel("Decay Constants")
plt.grid(True)

plt.tight_layout()
plt.show()
```
