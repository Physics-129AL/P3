```python
# Problem 3 A

import os
import numpy as np
import matplotlib.pyplot as plt

input_dir = 'Local_density_of_states_near_band_edge'
output_dir = 'Local_density_of_states_heatmap'

os.makedirs(output_dir, exist_ok=True)

files = [f for f in os.listdir(input_dir) if f.endswith('.txt')]

for file in files:
    with open(os.path.join(input_dir, file), 'r') as f:
        lines = f.readlines()
        data = [list(map(float, line.strip().split(','))) for line in lines]
    
    plt.figure(figsize=(8, 8))
    plt.imshow(data, cmap='hot', origin='lower', extent=[0, len(data[0]), 0, len(data)])
    plt.colorbar(label='Intensity')
    plt.title(f'Local Density of States - File Index {file[:-4]}')
    
    output_file = os.path.join(output_dir, f'{file[:-4]}_heatmap.png')
    plt.savefig(output_file)
    plt.close()

print(f'Heatmaps saved in {output_dir}.')
```

```python
# 3 B

import os
import numpy as np
import matplotlib.pyplot as plt

input_dir = 'Local_density_of_states_near_band_edge'
output_dir = 'Local_density_of_states_near_band_edge/local_density_of_states_height'

os.makedirs(output_dir, exist_ok=True)

files = [f for f in os.listdir(input_dir) if f.endswith('.txt')]

for file in files:
    with open(os.path.join(input_dir, file), 'r') as f:
        lines = f.readlines()
        data = np.array([list(map(float, line.strip().split(','))) for line in lines])
    
    x = np.arange(len(data[0]))
    y = np.arange(len(data))
    X, Y = np.meshgrid(x, y)
    
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(projection='3d')
    ax.plot_surface(X, Y, data, cmap='viridis')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Intensity')
    plt.title(f'Local Density of States - File Index {file[:-4]}')
    
    output_file = os.path.join(output_dir, f'{file[:-4]}_surface.png')
    plt.savefig(output_file)
    plt.close()

print(f'Surface plots saved in {output_dir}.')
```

```python
# 3 C
import os
import numpy as np
import matplotlib.pyplot as plt

input_dir = 'Local_density_of_states_near_band_edge'
subregion_coords = (50, 100, 50, 100)  
averages = []

files = [f for f in os.listdir(input_dir) if f.endswith('.txt')]

for file in files:
    with open(os.path.join(input_dir, file), 'r') as f:
        lines = f.readlines()
        data = np.array([list(map(float, line.strip().split(','))) for line in lines])
    
    x_start, x_end, y_start, y_end = subregion_coords
    subregion_data = data[y_start:y_end, x_start:x_end]
    
    avg_density = np.mean(subregion_data)
    averages.append(avg_density)

plt.figure(figsize=(10, 6))
plt.plot(range(len(files)), averages, marker='o')
plt.xticks(range(len(files)), range(len(files)))
plt.xlabel('State')
plt.ylabel('Average Local Density of States')
plt.title('Average Local Density of States in Sub-Region')
plt.tight_layout()

output_file = 'Local_density_of_states_near_band_edge/average_density_plot.png'
plt.savefig(output_file)
plt.show()

print(f'Average density plot saved in {output_file}.')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(averages, fill=True, color='skyblue')

plt.xlabel('Average Density')
plt.ylabel('Density')
plt.title('Kernel Density Estimation (KDE) Plot for Sub-Region')

plt.show()
```

```python
from scipy.stats import norm
from scipy.stats import mode

mode_result = mode(averages)
print(f"The mode of the data is: {mode_result.mode[0]}")

mean = np.mean(averages)
median = np.median(averages)

print(f"The mean of the data is: {mean}")
print(f"The median of the data is: {median}")

std_dev = np.std(averages)
print(f"The standard deviation of the data is: {std_dev}")

from scipy.stats import t

confidence_level = 0.95
confidence_interval = t.interval(confidence_level, len(averages)-1, loc=mean, scale=std_dev/np.sqrt(len(averages)))

print(f"The {confidence_level*100}% confidence interval for the mean is: ({confidence_interval[0]}, {confidence_interval[1]})")

point_of_interest = mean
pdf_at_point = norm.pdf(point_of_interest, loc=mean, scale=std_dev)

print(f"The probability density at {point_of_interest} is: {pdf_at_point}")
```
